```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Counting unique id's
unique_id=len(pd.unique(df["Id"]))
print("Number of Unique-ID: "+str(unique_id))
```

```python
### Conversion of ActivityDate dtype to datetime64 and format to yyyy-mm-dd
df["ActivityDate"] = pd.to_datetime(df["ActivityDate"],
                                      format="%m/%d/%Y")

#pre-printing information to confirm
df.info()
```

```python
# Printing the first 5 rows of "ACtivityDate" to confirm
df["ActivityDate"].head()
```

```python
# Creating new list of rearranged columns
new_cols = ['Id', 'ActivityDate', 'DayOfTheWeek', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalExerciseMinutes', 'TotalExerciseHours', 'Calories']
```

```python
#Reindexing columns based on new columns
df=df.reindex(columns=new_cols)
```

```python
# Printing 1st five rows to confirm
df.head()
```

```python
# create new column "day_of_the_week" to represent day of the week
df["DayOfTheWeek"] = df["ActivityDate"].dt.day_name()

# print 1st 5 rows to confirm
df["DayOfTheWeek"].head()
```

```python
# Renaming column names
df.rename(columns = {"Id":"Id", "ActivityDate":"Date", "DayOfTheWeek":"Day", "TotalSteps":"Total_Steps", "TotalDistance":"Total-Dist", "TrackerDistance":"Tracker-Dist", "LoggedActivitiesDistance":"Logged-Dist", "VeryActiveDistance":"Very-Active-Dist", "ModeratelyActiveDistance":"Moderate-Active-Dist", "LightActiveDistance":"Light-Active-Dist", "SedentaryActiveDistance":"Sedentary-Active-Dist", "VeryActiveMinutes":"Very-Active-Mins", "FairlyActiveMinutes":"Fairly-Active-Mins", "LightlyActiveMinutes":"Lightly-Active-Mins", "SedentaryMinutes":"Sedentary-Mins", "TotalExerciseMinutes":"Total-Mins","TotalExerciseHours":"Total-Hours","Calories":"Cals"}, inplace = True)
```

```python
# Re-printing columns to confirm
df.columns
```

```python
# Re-printing 1st five values of dataframe to confirm
df.head()
```

```python
# Creating new column "Total-Mins" containing sum of total minutes
df["Total-Mins"] = df["Very-Active-Mins"] + df["Fairly-Active-Mins"] + df["Lightly-Active-Mins"] + df["Sedentary-Mins"]

# Printing to confirm
df["Total-Mins"].head()
```

```python
#Re-printing 1st five values of df to confirm
df.head()
```

```python
# Creating a new column "Total-Hours" by converting the Total-Mins to Total-Hours
df["Total-Hours"]=round(df["Total-Mins"]/60)

#printing 1st five rows to confirm
df["Total-Hours"].head()
```

```python
#Re-printing 1st five values of df to confirm
df.head()
```

```python
# General Statistics
df.describe()
```

```python
# Relationship between calories burned and total steps walked in a day

figure = px.scatter(data_frame = df, x="Cals",
                    y="Total_Steps", size="Very-Active-Mins",
                    trendline="ols",
                    title="Relationship between Calories & Total Steps")
figure.show()
```

```python
# Calculating total of individual minutes column
very_active_mins = df["Very-Active-Mins"].sum()
fairly_active_mins = df["Fairly-Active-Mins"].sum()
lightly_active_mins = df["Lightly-Active-Mins"].sum()
sedentary_mins = df["Sedentary-Mins"].sum()

# plotting pie chart
slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ["Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary minutes"]
colours = ["#FFBBBB", "#1EAE98", "#6FD6FF", "#CD298A"]
explode = [0, 0, 0, 0.1]
plt.style.use("default")
plt.pie(slices, labels = labels,
        colors = colours, wedgeprops = {"edgecolor": "black"},
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go  # Make sure to import the necessary module

# Plotting graph for very active, fairly active, and lightly active minutes on each day of the week.
fig = go.Figure()

# Very Active minutes bar
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["Very-Active-Mins"],
    name='Very Active',
    marker_color='#1DE4BD'  # Color for very active
))

# Fairly Active minutes bar
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["Fairly-Active-Mins"],
    name='Fairly Active',
    marker_color='#003f5c'  # Color for fairly active
))

# Lightly Active minutes bar
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["Lightly-Active-Mins"],
    name='Lightly Active',
    marker_color='#EABD38'  # Color for lightly active
))

# Customize layout, set barmode to group for grouped bars, and adjust x-axis labels
fig.update_layout(barmode='group', xaxis_tickangle=-45)

# Display the figure
fig.show()
```

```python
# Plotting graph for No.of times users logged in app across the week

plt.style.use("default")
plt.figure(figsize=(6,4)) # specify size of the chart
plt.hist(df.Day, bins = 7,
         width = 0.6, color = "#bc5090", edgecolor = "black")

plt.xlabel("Day of the week")
plt.ylabel("Frequency")
plt.title("No. of times users logged in app across the week")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: Replace df.Total_Steps and df.Cals with your DataFrame columns
# df = your DataFrame here, with 'Total_Steps' and 'Cals' as relevant columns.

plt.style.use("default")
plt.figure(figsize=(8,6))  # specify size of the chart

# Scatter plot showing calories burned for each step taken
plt.scatter(df.Total_Steps, df.Cals,
            alpha = 0.8, c = df.Cals,
            cmap = "Spectral", edgecolor='k')

# Add median lines for steps and calories burned
median_calories = 2303  # Example value for median calories (replace with real median)
median_steps = 7637     # Example value for median steps (replace with real median)

# Adding color bar
plt.colorbar(orientation="vertical", label="Calories Burned")

# Median lines to show benchmarks
plt.axvline(median_steps, color="blue", linestyle="--", label="Median steps: 7637")
plt.axhline(median_calories, color="red", linestyle="--", label="Median calories: 2303")

# Labels and title for the plot
plt.xlabel("Steps Taken", fontsize=12)
plt.ylabel("Calories Burned", fontsize=12)
plt.title("Medical Fitness Curve: Calories Burned vs. Steps Taken", fontsize=14)

# Add a grid, legend, and show the plot
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Sample data: Replace df.Total_Steps and df.Cals with your DataFrame columns
# df = your DataFrame here, with 'Total_Steps' and 'Cals' as relevant columns.

# Extract the relevant features for clustering
X = df[['Total_Steps', 'Cals']]

# Apply KMeans clustering with 3 clusters (or any number you prefer)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Plotting the clustered scatter plot
plt.figure(figsize=(8,6))  # specify size of the chart

# Scatter plot for each cluster with different colors
plt.scatter(df.Total_Steps, df.Cals,
            c=df['Cluster'], cmap='viridis',
            alpha=0.8, edgecolor='k')

# Add cluster centers to the plot
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, label='Centroids')

# Adding median lines for steps and calories burned
median_calories = 2303  # Example value for median calories
median_steps = 7637     # Example value for median steps

plt.axvline(median_steps, color="blue", linestyle="--", label="Median steps: 7637")
plt.axhline(median_calories, color="red", linestyle="--", label="Median calories: 2303")

# Labels and title for the plot
plt.xlabel("Steps Taken", fontsize=12)
plt.ylabel("Calories Burned", fontsize=12)
plt.title("Clustered Medical Fitness Curve: Calories Burned vs. Steps Taken", fontsize=14)

# Add a grid, legend, and show the plot
plt.colorbar(label='Cluster')
plt.grid(True)
plt.legend()
plt.show()
```
